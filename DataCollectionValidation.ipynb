```python
import os
import re

# Directory where the images are stored
directory = '/home/hindupur.v/varsha_hpp/dataset/lung_colon_image_set/colon_aca'

# Regex pattern to match files like 'colonca1.jpeg'
pattern = re.compile(r'colonca(\d+)\.jpeg')

# List all files in the directory and filter using the regex pattern
files = os.listdir(directory)
colonca_files = [f for f in files if pattern.match(f)]

# Extract the numbers from these filenames and sort them
numbers = [int(pattern.search(f).group(1)) for f in colonca_files]
numbers.sort()

# The highest number should correspond to the last image uploaded
last_image_number = numbers[-1] if numbers else None
last_image_number
```

```python
import os

directory = '/home/hindupur.v/varsha_hpp/dataset/colon_aca'

def check_images_capture_missing(directory, base_image_name, count):
    missing_images=[]
    for i in range(1, count+1):
        image_name = f"{base_image_name}{i}.jpeg"
        file_path = os.path.join(directory, image_name)
        if not os.path.isfile(file_path):
            missing_images.append(i)
        return missing_images

base_image_name = 'colonca'
missing1 = check_images_capture_missing(directory, base_image_name, 5000)
missing1
```

```python
directory_path = '/home/hindupur.v/varsha_hpp/dataset/lung_colon_image_set/colon_aca'
filename = 'colonca'

# Generate a set of expected file names
expected_files = {f'{filename}{i}.jpeg' for i in range(1, 5001)}

# Get a set of actual file names in the directory
actual_files = set(os.listdir(directory_path))

# Find the difference - files that are expected but not found
missing_files = expected_files - actual_files

# Print missing files
if missing_files:
    print("Missing files:")
    # Correct the slicing based on the 'colonn' prefix and '.jpeg' extension
    for file in sorted(missing_files, key=lambda x: int(x[len(filename): -5])):
        print(file)
else:
    print("No files are missing.")
```
